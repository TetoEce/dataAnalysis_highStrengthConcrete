```python
y=data[data["Age (day)"]==28]
z=data[data["Superplasticizer percentage"]>0]
m=data[data["Fly Ash/Water"]>0]
e=pd.merge(y,z)
o=pd.merge(y,m)
dg=(pd.merge(e,o))
```

```python
dg
```

```python
#Reshaping
X=dg[["Concrete compressive strength(MPa, megapascals) "]]
Y=dg[["Superplasticizer percentage" and "Fly Ash/Water"]]
X=X.values.reshape(len(X),1)
Y=Y.values.reshape(len(Y),1)
print("I have reshaped x to:  ",X.shape)
print("I have reshaped y to:  ",Y.shape)
#Regression
reg=linear_model.LinearRegression()
reg=reg.fit(X,Y)
y_pr=reg.predict(Y)
print("real values of x: ",X)
print("real values of y: ",Y)
print("predicted values of y: ",y_pr)
print ("slopes of regression equation: ",round(reg.coef_[0][0],5), round(reg.coef_[0][0],4))
print ("intercept of regression equation: ",round(reg.intercept_[0],3))
print("Equation of the model is:","Compressive strength =",round(reg.coef_[0][0],5),"Superplasticizer",round(reg.coef_[0][0],4),"Flyash/water + ",round(reg.intercept_[0],3))
```
